```python
# Imports principaux pour l'analyse exploratoire
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# %matplotlib inline  # Permet d'afficher les graphiques directement dans le notebook au lieu d'une fenêtre separer.
%matplotlib inline


# ---------------------------------------------
# ETAPE 1 : ANALYSE EXPLORATOIRE DES BATIMENTS
# ---------------------------------------------

print("Étape 1 : Début de l'analyse exploratoire\n")

# 1. Chargement du fichier CSV
# On indique le chemin du fichier. Ici, il est dans le dossier 'data' à la racine du projet.
fichier_csv = "../data/2016_Building_Energy_Benchmarking.csv"

# On lit le CSV avec pandas pour obtenir un DataFrame, qui est une structure de données similaire à une table Excel
batiments = pd.read_csv(fichier_csv)
print(f"CSV chargé depuis : {fichier_csv}")
print(f"Nombre de bâtiments total : {batiments.shape[0]}")
print("Colonnes disponibles :", batiments.columns.tolist())
display(batiments.head(3))  # Affiche les 3 premières lignes pour vérifier

# 2. Aperçu initial des données
# .head() affiche les 5 premières lignes du tableau pour avoir une idée des colonnes et valeurs
# .info() permet de vérifier le nombre de lignes, les types de colonnes et s'il y a des valeurs manquantes
print("\nInformations générales sur le jeu de données :")
batiments.info()

# 3. Sélection des colonnes pertinentes
# On ne conserve que les colonnes utiles pour l'analyse et la modélisation
colonnes_pertinentes = [
    'PropertyName',             
    'PrimaryPropertyType',     
    'Address',                  
    'YearBuilt',                
    'PropertyGFABuilding(s)',  
    'SiteEUI(kBtu/sf)',        
    'SiteEnergyUse(kBtu)',     
    'TotalGHGEmissions',    
    'Latitude',            
    'Longitude',
    'Neighborhood',
    'ListOfAllPropertyUseTypes',
    'Electricity(kBtu)',
    'NaturalGas(kBtu)',
    'SteamUse(kBtu)'
]
batiments = batiments[colonnes_pertinentes]
print("\nColonnes conservées :", batiments.columns.tolist())
display(batiments.head(3))  # Vérification après sélection


# 4. Définition des types utiles basés sur la fréquence observée dans les données
# Ici, nous sélectionnons les catégories qui ont assez d'échantillons (70+) pour être utiles
# dans une modélisation fiable. Les catégories trop rares ou trop générales sont exclues.
types_utiles = [
    'Low-Rise Multifamily',        # Très fréquent : données nombreuses et homogènes
    'Mid-Rise Multifamily',        # Très fréquent : cohérent pour l'analyse
    'High-Rise Multifamily',       # Suffisamment fréquent et homogène
    'Small- and Mid-Sized Office', # Catégorie bien représentée et cohérente
    'Large Office',                # Bureaux de grande taille, assez fréquents
    'Warehouse',                   # Nombre suffisant pour analyse (entrepôts)
    'Retail Store',                # Magasins, assez nombreux pour être utilisables
    'Hotel',                       # Catégorie homogène avec nombre acceptable
    'K-12 School'                  # Écoles, assez de données et cohérentes
]

# Affichage des types utiles sélectionnés pour confirmation
print("Types utiles sélectionnés pour le filtrage :")
print(types_utiles)
print("\n---------------------------------------------\n")

# Nombre de bâtiments avant filtrage
# .shape[0] donne le nombre total de lignes (bâtiments) dans le DataFrame
avant = batiments.shape[0]
print(f"Nombre de bâtiments AVANT filtrage : {avant}")

# Filtrage des bâtiments selon les types utiles
# isin() vérifie si la valeur de PrimaryPropertyType appartient à la liste types_utiles
# Le DataFrame ne garde que les bâtiments dont le type est dans la liste
batiments = batiments[batiments['PrimaryPropertyType'].isin(types_utiles)]

# Nombre de bâtiments après filtrage
apres = batiments.shape[0]
print(f"Nombre de bâtiments APRÈS filtrage : {apres}")
print(f"Bâtiments conservés : {apres} sur {avant}")
print("\n---------------------------------------------")



# 5. Filtrage des bâtiments aberrants
# On supprime les bâtiments avec des valeurs impossibles ou incohérentes
avant_filtrage = batiments.shape[0]
batiments = batiments[(batiments['PropertyGFABuilding(s)'] > 0) &
                      (batiments['SiteEUI(kBtu/sf)'] > 0) &
                      (batiments['SiteEnergyUse(kBtu)'] > 0)]
apres_filtrage = batiments.shape[0]
print(f"\nNombre de bâtiments après filtrage : {apres_filtrage} (avant : {avant_filtrage})")
display(batiments.head(3))  # Aperçu après filtrage

# 6. Vérification des valeurs manquantes
# On compte combien de valeurs sont manquantes dans chaque colonne
#Detection des valeurs manquantes avec .isna() il crée un tableau de True/False qui dit si chaque valeur est :
#True  manque
#False  non manquante
print("\nNombre de valeurs manquantes par colonne :")
display(batiments.isna().sum())

# 7. Choix de la target
# Pour la suite du projet, il faut choisir **une seule colonne à prédire**.
# Ici on choisit 'SiteEnergyUse(kBtu)', car elle représente la consommation totale d’énergie du bâtiment
target = 'SiteEnergyUse(kBtu)'
print(f"\nTarget choisie pour la modélisation : {target}")

# 8. Statistiques descriptives
# On regarde la moyenne, le min, max, quartiles, etc., de la target
print("\nStatistiques descriptives de la target :")
display(batiments[target].describe())

# 9. Visualisation : distribution de la target
# On utilise un histogramme pour voir comment se répartissent les consommations
plt.figure(figsize=(8,5))
sns.histplot(batiments[target], bins=30, kde=True)  # kde= en True ajoute une courbe de densité
plt.xlabel('Consommation d\'énergie (kBtu/ft²)')
plt.ylabel('Nombre de bâtiments')
plt.title('Distribution de la consommation d\'énergie')
plt.show()

# 10. Visualisation : consommation selon le type de bâtiment
# Boxplot : pratique pour voir les valeurs extrêmes et comparer plusieurs catégories
plt.figure(figsize=(12,6))
sns.boxplot(x='PrimaryPropertyType', y=target, data=batiments)
plt.xticks(rotation=45, ha='right')
plt.title('Consommation d\'énergie par type de bâtiment')
plt.show()

# 11. Observations préliminaires
print("\nObservations préliminaires :")
print("- Vérifier quels types de bâtiments sont les plus fréquents.")
print("- Repérer les bâtiments avec des consommations très élevées ou très faibles (outliers).")
print("- Les colonnes peuvent contenir encore des valeurs manquantes, à traiter avec prudence pour la modélisation.")
print("- Toujours garder une trace du nombre de lignes avant/après nettoyage pour justifier le filtrage.")
```

```python
# ---------------------------------------------
# ETAPE 2 : FEATURE ENGINEERING
# ---------------------------------------------

print("\nÉtape 2 : Feature Engineering\n")

# 1️⃣ Création de features temporelles
# On peut extraire l'âge du bâtiment à partir de l'année de construction
batiments['BuildingAge'] = 2016 - batiments['YearBuilt']  # Age du bâtiment
print("Feature 'BuildingAge' ajoutée : âge du bâtiment calculé à partir de YearBuilt")
display(batiments[['YearBuilt','BuildingAge']].head(3))

# 2️⃣ Features sur la structure du bâtiment
# On utilise PropertyGFABuilding(s) pour normaliser ou créer des ratios si besoin
batiments['BuildingSize'] = batiments['PropertyGFABuilding(s)']  # Surface totale du/des bâtiments
print("Feature 'BuildingSize' ajoutée : même valeur que PropertyGFABuilding(s) pour clarté et futur usage")
display(batiments[['PropertyGFABuilding(s)','BuildingSize']].head(3))

# 3️⃣ Features binaires sur types de bâtiments
# Pour chaque type utile, on crée une colonne 0/1 indiquant si le bâtiment est de ce type
types_utiles = [
    'Hotel',
    'K-12 School',
    'Small- and Mid-Sized Office',
    'Large Office',
    'Mixed Use Property',
    'Medical Office',
    'Senior Care Community',
    'Retail Store',
    'University'
]

for t in types_utiles:
    batiments[f'Is_{t.replace(" ","_").replace("-","_")}'] = (batiments['PrimaryPropertyType'] == t).astype(int)
    print(f"Feature binaire 'Is_{t}' créée : 1 si type {t}, 0 sinon")

display(batiments[['PrimaryPropertyType'] + [f'Is_{t.replace(" ","_").replace("-","_")}' for t in types_utiles]].head(5))

# 4️⃣ Features géographiques
# On peut ajouter la latitude et longitude directement, ou créer des zones (quartiers)
batiments['Latitude'] = batiments['Latitude'].astype(float)
batiments['Longitude'] = batiments['Longitude'].astype(float)
print("Features 'Latitude' et 'Longitude' conservées pour localisation")

# 5️⃣ Features sur usage multiple
# ListOfAllPropertyUseTypes peut indiquer combien d'usages différents pour chaque bâtiment
# On crée une feature "NbPropertyUses" qui compte le nombre de types d'usage
batiments['NbPropertyUses'] = batiments['ListOfAllPropertyUseTypes'].fillna('').apply(lambda x: len(x.split(',')) if x != '' else 1)
print("Feature 'NbPropertyUses' ajoutée : nombre de types d'usages différents dans le bâtiment")
display(batiments[['ListOfAllPropertyUseTypes','NbPropertyUses']].head(5))

# 6️⃣ Features sur sources d'énergie (sans leakage)
# On ne met pas les valeurs consommées, mais juste l'existence de chaque type
batiments['HasElectricity'] = (batiments['Electricity(kBtu)'] > 0).astype(int)
batiments['HasNaturalGas'] = (batiments['NaturalGas(kBtu)'] > 0).astype(int)
batiments['HasSteam'] = (batiments['SteamUse(kBtu)'] > 0).astype(int)
print("Features 'HasElectricity', 'HasNaturalGas', 'HasSteam' ajoutées : indiquent la présence de chaque source d'énergie sans fuite de données")

display(batiments[['Electricity(kBtu)','NaturalGas(kBtu)','SteamUse(kBtu)','HasElectricity','HasNaturalGas','HasSteam']].head(3))

# 7️⃣ Vérification finale des features
print("\nAperçu final des nouvelles features :")
display(batiments.head(3))

# 8️⃣ Statistiques descriptives pour les features numériques
print("\nStatistiques descriptives des features numériques :")
display(batiments[['BuildingAge','BuildingSize','NbPropertyUses','HasElectricity','HasNaturalGas','HasSteam']].describe())

# ✅ Feature engineering terminé
print("\nFeature Engineering : terminé. Toutes les features principales sont créées et prêtes pour la modélisation.")
```
